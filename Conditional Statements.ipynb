## `if`, `elif`, `else` in Python 

### `if`

* Used to **check a condition**
* Executes when the condition is **True**

### `elif`

* Stands for **else if**
* Used to check **multiple conditions**
* Checked **only if previous condition is False**
* Can be used **multiple times**

### `else`

* Executes when **all above conditions are False**
* Comes **at the end**
* Used as a **default block**


### Important Points

* Conditions are evaluated **top to bottom**
* Only **one block executes**
* Indentation is **mandatory** in Python
* `elif` and `else` are **optional**
* Used for **decision-making logic**

### Use Cases

* Grading systems
* Login validation
* Age checks
* Discount rules
* Decision-based programs


```python
# if else 
# syntax 
# if condition:
#     statement
# else:
#     statement 
    
age = 144

if age>=18:
    print('you can vote')
else: 
    print('you cannot vote')
```

```python
# Data validations

marks = 99

if marks>=0 and marks<101:
    print('valid marks')
else:
    print('invalid marks')
```

```python
# even-odd 
n1 = -2
if n1%2==0:
    print('even')
else:
    print('odd')
```

```python
# elif cases 
# multiple conditions

# check number if it is positive or negative 

n1 = 0

if n1>0:
    print('positive')
elif n1<0:
    print('negative')
else:
    print('neutral')
```

```python
# nested if else 
# password authentication

user = 'anshumbanga'
password = '123456'

u1 = input('enter your username')

if u1==user: # true 
    print('valid username')
    print('-'*50)
    keys = input('enter your password')
    if keys==password:
        print('access granted')
    else:
         print('Incorrect password')
else:
    print('invalid username')
```
