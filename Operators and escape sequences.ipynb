# operators
Operators in Python are symbols or keywords used to perform operations on values or variables (called operands).

# arithemetic operators 
Arithmetic operators in Python are used to perform mathematical calculations.

```python
+ : Addition
- : subtraction
* : multiplication
/ : division
//: floor division
% : modulus(remainder)
**: exponential
```

```python
a = 10 
b = 20 

a+b
```

```python
5%2 # remainder 
47%7 # remainder 

# ** 

2**10 # exponential 

# floor division
5//2
```

# assignment operators
Assignment operators in Python are used to assign values to variables and sometimes perform an operation while assigning.

```python
= → assigns a value
+= → adds and assigns
-= → subtracts and assigns
*= → multiplies and assigns
/= → divides and assigns
%= → modulus and assigns
```

```python
a = 10 
a
```

```python
a + 10 # operation
a * 10 
```

```python
# change value 
# a = a*100 # overwrite
a*=100 # multiply and assign
```

```python
a
```

# comparision operators
Comparison operators in Python are used to compare two values and return a Boolean result (True or False).

```python
== → equal to
!= → not equal to
> → greater than
< → less than
>= → greater than or equal to
<= → less than or equal to
```

```python
10>5
-1<-10
a = 10 # assign 
a == 10 # compare
12 == 20 # equals to
12 != 20 # not equals to
```

# in operator
in operator in Python is used to check whether a value exists inside a sequence.

```python
# str 
str1 = 'iqra'

'i' in str1
'z' in str1
'I' in str1
```

# logical operators 
Logical operators in Python are used to combine or modify conditions and return a Boolean result (True or False).

```python
and → True if both conditions are True
or → True if at least one condition is True
not → Reverses the result
```

```python
True and True and False
True or False
False and True
10>2 and 12>5
10<15 and 10<8
not True
not False
```

# escape Sequences
Escape sequences in Python are special characters used inside strings, written with a backslash (\), to represent actions instead of normal text.

\n → New line

```python
print('hello\nworld')
```

```python
name = 'jeet\nkumar'
name
```

```python
print(name) # print statements
```

\t → Tab space

```python
# formatting
print('name:\tiqra')
print('course:\tDS')
print('age:\t24')
```

\' → Single quote

```python
str1 = 'jeet doesn\'t likes to study'
str1
```

\" → Double quote

```python
str2 = "jeet said,\"I am not comming today\"."
str2
```

\\ → Backslash

```python
print('this is a backslash \\')
```
